```python
import numpy as np
import matplotlib.pylab as plt # this library is used for reading images
def plti(im, h=10, **kwargs):
    # Helper function to plot an image.
    y = im.shape[0]
    x = im.shape[1]
    w = (y/x) * h
    plt.figure(figsize=(w,h))
    plt.imshow(im, interpolation="none", **kwargs)
    plt.axis('off')    
    # input('press <ENTER> to continue')

%matplotlib inline
im = plt.imread("./../Images/airplane.jpg")
print(type(im))
# plti(im)
# im = im[400:3800,:2000,:]    
# plti(im)
print(im.shape)
print(im[:5, :10, :10])

# im = im[400:3800,:2000,:]    
# plti(im)
plti(im)
```
